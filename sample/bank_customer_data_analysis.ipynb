```python
ax2=sns.countplot(x="education",data=data)
plt.title("Education type vs Count")
# plt.hist(data['education'],bins=4,edgecolor='red')
size=10
for p in ax2.patches:
   ax2.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.11), fontsize=9,rotation=0)

plt.show()
```

```python
data.drop('id', axis=1, inplace=True)
encoded_job = pd.get_dummies(data['job'])

# Concatenate the encoded columns back to the original DataFrame
df = pd.concat([data, encoded_job], axis=1)

# Drop the original 'job' column
df.drop('job', axis=1, inplace=True)
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data.drop('y', axis=1), data['y'], test_size=0.2, random_state=42)

# Train the logistic regression model
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

# Predict on the test set
y_pred = logreg.predict(X_test)

# Calculate accuracy
accuracy = logreg.score(X_test, y_test)
print('Accuracy:', accuracy)
```
